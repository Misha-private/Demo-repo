### Plotting times of v1 and v2 schemes

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({'Processes': ['an2filt', 'intpT', 'vfiltT', 'upsend', 'hfiltT',
                                 'ssendT', 'weight', 'ssend', 'hfilt', 'dnsend', 'vfilt', 'intp',
                                 'filt2an'],
                   'Times':     [0.272, 0.216, 0.1095, 0.198, 0.916, 0.184, 0.031, 0.157,
                                 0.531, 0.312, 0.092, 0.126, 0.166]})

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

major_tick = np.arange(0, 1.2, 0.1)
minor_tick = np.arange(0, 1.2, 0.05)
ax.set_yticks(major_tick) # Grid
ax.set_yticks(minor_tick, minor=True)

#ax.grid(which='both')

#ax.grid(which='minor', alpha=1)
#ax.grid(which='major', alpha=2)

sns.set_theme(style='whitegrid')


plt.title('Times in test with $v_1$ scheme')
plt.grid(which='both',axis='y',linestyle='--', alpha=0.8)
sns.barplot(x='Processes', y='Times', color='#DC3220', gap=0.3, data=df)
plt.xticks(rotation=45)
plt.ylim(0., 1.)

plt.savefig('times_v1.png')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df = pd.DataFrame({'Processes': ['an2filt', 'intpT', 'vfiltT', 'upsend', 'hfiltT',
                                 'ssendT', 'weight', 'ssend', 'hfilt', 'dnsend', 'vfilt', 'intp',
                                 'filt2an'],
                   'Times':     [0.271, 0.216, 0.110, 0.208, 0.336, 0.305, 0.032, 0.303, 0.270,
                                 0.310, 0.091, 0.126, 0.167]})

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

major_tick = np.arange(0, 1.2, 0.1)
minor_tick = np.arange(0, 1.2, 0.05)
ax.set_yticks(major_tick) # Grid
ax.set_yticks(minor_tick, minor=True)

#ax.grid(which='both')

#ax.grid(which='minor', alpha=1)
#ax.grid(which='major', alpha=2)

sns.set_theme(style='whitegrid')


plt.title('Times in test with $v_2$ scheme')
plt.grid(which='both',axis='y',linestyle='--', alpha=0.8)
sns.barplot(x='Processes', y='Times', color='#005AB5', gap=0.3, data=df)
plt.xticks(rotation=45)
plt.ylim(0., 1.)

plt.savefig('times_v2.png')
plt.show()
```
